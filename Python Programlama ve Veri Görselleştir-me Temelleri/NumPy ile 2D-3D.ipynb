```python
#NumPy ile 2D-3D matris üzerinden görüntü temsilini anlamak
#2D
import numpy as np
import matplotlib.pyplot as plt

# 5x5 piksel gri tonlamalı görüntü (rastgele değerlerle)
gri_goruntu = np.array([
    [0, 50, 100, 150, 200],
    [10, 60, 110, 160, 210],
    [20, 70, 120, 170, 220],
    [30, 80, 130, 180, 230],
    [40, 90, 140, 190, 240]
], dtype=np.uint8)

plt.imshow(gri_goruntu, cmap='gray')
plt.title("2D Gri Tonlama Görüntüsü")
plt.colorbar()
plt.show()
```

```python
# 3D
rgb_goruntu = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],        # Kırmızı, Yeşil, Mavi
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],  # Sarı, Camgöbeği, Mor
    [[128, 128, 128], [255, 255, 255], [0, 0, 0]]   # Gri, Beyaz, Siyah
], dtype=np.uint8)

# Görüntüyü çizdir
plt.imshow(rgb_goruntu)
plt.title("3D Renkli (RGB) Görüntü")
plt.axis('off')
plt.show()
```
