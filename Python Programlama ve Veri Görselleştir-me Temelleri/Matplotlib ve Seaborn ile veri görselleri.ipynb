```python
#Matplotlib üzerinden  
import numpy as np
import matplotlib.pyplot as plt

# Rastgele 100x100 RGB görüntü
resim = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)

renkler = ['Red', 'Green', 'Blue']
for i, renk in enumerate(renkler):
    plt.hist(resim[:, :, i].ravel(), bins=256, color=renk.lower(), alpha=0.5, label=renk)

plt.title("RGB Histogram (Matplotlib)")
plt.xlabel("Piksel Değeri")
plt.ylabel("Frekans")
plt.legend()
plt.show()
```

```python
veri = np.random.rand(10, 10)

plt.imshow(veri, cmap='hot')
plt.title("Heatmap (Matplotlib)")
plt.colorbar()
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='teal')
plt.title("Scatter Plot (Matplotlib)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
veriler = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(veriler, patch_artist=True)
plt.title("Boxplot (Matplotlib)")
plt.xticks([1, 2, 3], ['Grup 1', 'Grup 2', 'Grup 3'])
plt.show()
```

```python
# Matplotlib ve Seaborn ile RGB kanal bazlı histogram, heatmap, scatter, boxplot
import pandas as pd
import seaborn as sns
resim = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)
df = pd.DataFrame({
    'Red': resim[:, :, 0].ravel(),
    'Green': resim[:, :, 1].ravel(),
    'Blue': resim[:, :, 2].ravel()
})

for renk in ['Red', 'Green', 'Blue']:
    sns.histplot(df[renk], bins=256, color=renk.lower(), alpha=0.5, label=renk, kde=False)

plt.title("RGB Histogram (Seaborn)")
plt.legend()
plt.show()
```

```python
veri = np.random.rand(10, 10)
sns.heatmap(veri, cmap='YlGnBu', annot=True, fmt=".2f")
plt.title("Heatmap (Seaborn)")
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
df_scatter = pd.DataFrame({'X': x, 'Y': y})

sns.scatterplot(data=df_scatter, x='X', y='Y', color='purple')
plt.title("Scatter Plot (Seaborn)")
plt.show()
```

```python
veriler = [np.random.normal(0, std, 100) for std in range(1, 4)]
df_box = pd.DataFrame({
    'Grup 1': veriler[0],
    'Grup 2': veriler[1],
    'Grup 3': veriler[2]
})

sns.boxplot(data=df_box)
plt.title("Boxplot (Seaborn)")
plt.show()
```
