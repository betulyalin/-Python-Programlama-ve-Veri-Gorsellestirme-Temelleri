```python
# Basit örneeklerle Pandas ile tablo görüntüsü, missing data analizi, histogram oluşturma.

import pandas as pd

# Örnek veri seti
veri = {
    "İsim": ["Ahmet", "Mehmet", "Ayşe", "Fatma", "Ali"],
    "Yaş": [25, 30, None, 22, 28],
    "Boy": [175, 180, 165, None, 170],
    "Kilo": [70, 85, 55, 60, None]
}

df = pd.DataFrame(veri)

print(df)
```

```python
# Hangi sütunda kaç tane eksik değer var?
print(df.isnull().sum())

# Tüm tablo için eksik veri sayısı
print("\nToplam eksik değer sayısı:", df.isnull().sum().sum())
```
